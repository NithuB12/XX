```python
import pandas as pd
pd.set_option('display.max_columns',10)
print("Original Dataset from Excel file\n")
df=pd.read_excel('/content/excel 1.xlsx')
print(df)
```

```python
f=df['NAME'].str.strip()
f.to_excel('excel2.xlsx')
```

```python
print("Replace value with forward\n")
k=df.fillna(method='pad')
print(k)
print('The above updated data set will be stored in excel2.xlsx file\n')
k.to_excel('/content/excel2.xlsx')
```

```python
print("Drop one particular column and its values\n")
k.drop(['TOTAl'],axis=1,inplace=True)
print(k)
print("The above updated data set will be stored in excel3.xlsx file\n")
k.to_excel('excel3.xlsx')
```

```python
print("Drop NaN rows\n")
df=pd.read_excel('excel3.xlsx')
print(df)
x=df.dropna()
print(x)
print("The above updated data set will be stored in excel4.xlsx file\n")
x.to_excel('excel4.xlsx')
```

```python
print("Replace values\n")
n=pd.read_excel('/content/excel3.xlsx')
print("Original data set\n")
print(n)
y=n.replace({49:50})
print("Updated data set with replaced values:{49:50}\n")
print(y)
print("The above updated data set will be stored in excel5.xlsx file\n")
y.to_excel('excel5.xlsx')
```
