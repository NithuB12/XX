```python
import numpy as np
a=np.array([[1,2,4], [5,8,7]])
print("Array created using passed list:\n",a)
```

```python
#creating a 3x4 array with all zeros
b=np.zeros((3,4))
print("\n An array initialised with all zeroes:\n",b)
```

```python
#create a constant value array of complex type
c=np.full((3,3),6)
print("\n An array initialised with all 6s.\n",c)
```

```python
#create an array with random values
d=np.random.random((2,2))
print("\nA random array:\n",d)
```

```python
#create a sequence of integers from 0 to 30 with steps of 5
e=np.arange(0,30,5)
print("\nA sequential array with steps of 5:\n",e)
```

```python
#reshaping 3x4 array to 2x2x3 array
arr=np.array([[1,2,3,4],[5,2,4,2],[1,2,0,1]])
newarr=arr.reshape(4,3)
print("\nOriginal array:\n",arr)
print("Reshaped array[4,3]:\n",newarr)
```

```python
#flatten array
flarr=arr.flatten()
print("\nOriginal array;\n",arr)
print("\nFlattened array;\n",flarr)
```

```python
#printing array dimensions
print("\nNo. of dimensions:",arr.ndim)
```

```python
#printing shape of array
print("\nShape of array: ", arr.shape)
```

```python
#printing size (total no of elements) of array
print("\nSize of array: ", arr.size)
```

```python
#printing type of elements in array
print("\nArray stores element of type:", arr.dtype)
```

```python
#coverting datatypes from integer to float
new=arr.astype('f')
print("\nConverted array:\n",new)
print("Converted array type:",new.dtype)
```
