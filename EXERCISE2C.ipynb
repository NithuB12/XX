```python
import pandas as pd
import numpy as np

#create series from array
arr=np.array(['P','a','n','d','a','s'])
a=pd.Series(arr)
print('series from array:')
print(a)
```

```python
#series from dictionary
arr={'x':0.,'y':1.,'z':2.}
b=pd.Series(arr)
print("\nSeries from dictionary:\n")
print(b)
```

```python
#Series using scalar
c=pd.Series(4,index=[0,1,2,3])
print("\nSeries using scalar\n")
print(c)
```

```python
#series through index
d=pd.Series([1,2,3],index=['a','b','c'])
print("\nSeries through index:\n")
print(d)
```

```python
e=pd.Series(data=[1,2,3,4])
print("\nSeries:\n",e)
```

```python
print("\nIndex:\n",e.index)
```

```python
print("\nValues:\n",e.values)
```

```python
print("\nShape:\n",e.shape)
```

```python
print("\nDimension:\n",e.ndim)
```

```python
print("\nSize:\n",e.size)
```
