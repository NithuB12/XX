```python
import pandas as pd
import numpy as np
from datetime import datetime

ran=pd.date_range("2020-08-15","2020-09-01")
print("Data range:\n",ran)

ran=pd.date_range("2020-07-15",periods=10)
print("\nData range with periods and default day frequency:\n",ran)

ran=pd.date_range("2020-07-15",periods=10,freq='H')
print("\nData range with hours and frequency:\n",ran)

ran=pd.period_range("2020-10",periods=10,freq='M')
print("\nData range with periods and month frequency:\n",ran)

ran=pd.timedelta_range(0,periods=8,freq="H")
print("\nZero day with hour frequency:\n",ran)
```
