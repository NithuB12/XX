```python
import numpy as np
arr1=[10,20,30,40,50]
arr2=[2,4,5,8,10]
a=np.array(arr1)
b=np.array(arr2)
print("Original arrays:")
print(a)
print(b)
print("Vector addition:",a+b)
print("Vector subtraction:",a-b)
print("Vector multiplication:",a*b)
print("Vector division:",a/b)
print("Vector dot product:",a.dot(b))
print("Scalar multiplication:")
scalar=5
print("Scalar value:",scalar)
print("Array:",a)
print("Result:",a*scalar)

```

```python
import numpy as np
#numpy vectorise method
def myfunction(x,y):
  #return x-y ifx>y,otherwise return x+y
  if x>y:
    return x-y
  else:
     return x+y

print("\nNumpy vectorise method")
print("(Return x-y if x>y, otherwise return x+y)")
arr1=[10,4,20]
arr2=[2,3,30]
vector=np.vectorize(myfunction)
print("Array 1:",arr1)
print("Array 2:",arr2)
print("Result:",vector(arr1,arr2))
```
