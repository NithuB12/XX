```python
import pandas as pd
import numpy as np

#DataFrame.Apply function
print("DataFrame.applyfunction:\n")
a=pd.DataFrame([[2,7]]*3,columns=['P','Q'])
print("\nOriginal Data:\n",a)
```

```python
print("\nSquare root of Data Frame;\n",a.apply(np.sqrt))
```

```python
print("\nSum of each column:\n",a.apply(np.sum,axis=0))
```

```python
print("\nSum of each row:\n",a.apply(np.sum,axis=1))
```

```python
a=pd.DataFrame([[1,5,7],[2,7,8],[3,6,9]],columns=['X','Y','Z'])
print("\nOriginal Data Frame:\n",a)
```

```python
print("\nMinimum and Maximum of each column:\n")
print(a.agg(['min','max']))
```

```python
#DataFrame assign function
print("\nData Frame Assign Function:\n")
b=pd.DataFrame([['Dale',123],['Mark',143]],columns=['Employee','ID'])
print("\nOriginal Data Frame:\n",b)

b['Age']=[35,40]
print("\nAdding new column:\n",b)

c=b.assign(Sex=['Male','Male'])
print("\nAdding a new column:\n",c)
```

```python
#Data Frame Merge Function
print("\nData Frame Merge Function:\n")
left=pd.DataFrame({"ID":[1,2,3,4,5],'Name':['Alex','Amy','Allen','Alice','Ayoung'],'Sub':['Sub1','Sub2','Sub3','Sub4','Sub5']})
right=pd.DataFrame({'ID':[1,2,3,4,5],'Name':['Billy','Brian','Bran','Bryce','Betty'],'Sub':['Sub1','Sub2','Sub3','Sub4','Sub5']})
print(left)
print(right)
print(pd.merge(left,right,on='ID'))
```
